#### Seaborn :Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies the creation of complex visualizations such as heatmaps, violin plots, and pair plots, while maintaining fine-grained control over aesthetics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

```

#### Seabron has some inbulit functions

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
tips.head()
```

```python
# Summary statistics for numerical columns
print(tips.describe())
```

```python
# Check for missing values
print(tips.isnull().sum())
```

```python
# Count plot for 'smoker' column
sns.countplot(x='smoker', data=tips)
plt.title('Smoker vs Non-Smoker')
plt.show()
```

```python
# Count plot for 'sex'
sns.countplot(x='sex', data=tips)
plt.title('Male vs Female')
plt.show()

# Count plot for 'day'
sns.countplot(x='day', data=tips)
plt.title('Tips per Day')
plt.show()
```

```python
# Scatter plot: Total Bill vs Tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
# Calculate average tip by gender
average_tip_by_gender = tips.groupby('sex')['tip'].mean()
print(average_tip_by_gender)
```

```python
# Box plot: Tip amount by Day
sns.boxplot(x='day', y='tip', data=tips)
plt.title('Tip Distribution by Day')
plt.show()
```

```python
# Bar plot: Average Tip by Gender
sns.barplot(x='sex', y='tip', data=tips)
plt.title('Average Tip by Gender')
plt.show()
```

```python
# Correlation matrix
corr_matrix = tips.corr()

# Plot the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot: Shows pairwise relationships of numerical columns
sns.pairplot(tips)
plt.show()
```

```python
# FacetGrid: Create scatter plots of 'total_bill' vs 'tip' for each 'day'
g = sns.FacetGrid(tips, col='day')
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
# Violin plot: Tip distribution by Day
sns.violinplot(x='day', y='tip', data=tips)
plt.title('Tip Distribution by Day')
plt.show()
```

#### Explore the iris Dataset:

```python
# Display the first few rows of the iris dataset
print(iris.head())
```

```python
# Histogram for sepal_length
sns.histplot(iris['sepal_length'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
# Scatter plot of sepal_length vs petal_length
sns.scatterplot(x='sepal_length', y='petal_length', data=iris, hue='species')
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
# Pairplot: Explore relationships between all numerical columns
sns.pairplot(iris, hue='species')
plt.show()
```

```python
# Correlation heatmap for the iris dataset
corr_matrix = iris.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - Iris Dataset')
plt.show()
```

#### Explore the titanic Dataset: 

```python
# Display the first few rows of the titanic dataset
print(titanic.head())
```

```python
# Bar plot showing survival rate by sex
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
# Histogram of age distribution based on survival
sns.histplot(titanic[titanic['survived'] == 1]['age'], kde=True, color='green', label='Survived')
sns.histplot(titanic[titanic['survived'] == 0]['age'], kde=True, color='red', label='Not Survived')
plt.title('Age Distribution of Titanic Passengers')
plt.legend()
plt.show()
```

```python
# Correlation matrix for the Titanic dataset
corr_matrix = titanic.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - Titanic Dataset')
plt.show()
```

```python
# Box plot: Survival by Class
sns.boxplot(x='pclass', y='survived', data=titanic)
plt.title('Survival by Class')
plt.show()
```

#### 4. Explore the planets Dataset:

```python
# Display the first few rows of the planets dataset
print(planets.head())
```

```python
# Histogram for orbital_period
sns.histplot(planets['orbital_period'], kde=True)
plt.title('Distribution of Orbital Periods')
plt.show()
```

```python
# Scatter plot of mass vs orbital period
sns.scatterplot(x='mass', y='orbital_period', data=planets)
plt.title('Mass vs Orbital Period')
plt.show()
```

```python
# Count plot for the number of planets discovered each year
sns.countplot(x='year', data=planets)
plt.title('Number of Planets Discovered Each Year')
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation matrix for the planets dataset
corr_matrix = planets.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap - Planets Dataset')
plt.show()
```
