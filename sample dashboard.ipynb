```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
x =[1,2,3,4,5]
y =[1,4,9,16,25]
plt.figure(figsize=(5,4))
plt.plot(x,y, color="red", linestyle='--',marker='o', linewidth=2)
plt.grid()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Line Plot")
plt.show()
```

#### Subplot

```python
x =[1,2,3,4,5]
y1 =[1,4,9,16,25]
y2 =[1,2,3,4,5]
plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")
plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title("Plot 2")
plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title("Plot 3")
plt.subplot(2,2,4)
plt.plot(y1,x,color='green')
plt.title("Plot 4")
```
